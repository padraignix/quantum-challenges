## Part A: Design a VQC on an ideal backend

Variational quantum classifiers (VQCs) are a type of variational quantum algorithm that can be used to solve certain types of classification problems. In particular, VQCs feature a hybrid architecture. This architecture is based on an ansatz in the form of a parametrized quantum circuit applied onto an initial state. The output is measured in the form of a cost function, which is classically optimized over the circuit's parameters until we converge to a minimum.



In this Lab, you are given a list of 20 birds:

 > Falcon, Hummingbird, Eagle, Osprey, Condor, Heron, Peacock, Parrot, Swan, Toucan, Cardinal, Macaw, Puffin, Kestrel, Dove, Woodpecker, Sparrow, Stork, Crow, Seagull.

The first six birds of the list are found in the category of IBM-quantum birds, since processor types were named after them - check them [here](https://docs.quantum.ibm.com/run/processor-types). The rest of them are non-IBM quantum birds.

The dataset is given to you in the form of a dictionary, *bird_dataset.csv*. Each bird corresponds to an entry, and it is associated to a 5-qubit quantum state given in a statevector form. Therefore, for each bird you will find $2^5 = 32$ coefficients.

In particular, IBM-quantum birds are associated entangled quantum states, while non-IBM birds are associated to non-entangled (product) quantum states. Turns out that distinguishing between entangled and non-entangled quantum states is not so easy a priori! The goal of this part of the Lab is to build a VQC which is able to find the IBM-quantum subset of birds. ### QP step 1: Map classical inputs to a quantum problem

In this step of QPs, we are interested in learning how to translate a problem into a quantum circuit and a set of measurements that can reasonably run on quantum hardware. Note that this is in general a hard problem, so here's a list of things to take into account:

- We want to choose problems that are classically hard. Distinguishing between entangled and non-entangled states is a classically hard problem, even though we are choosing a small scale problem for our learning purposes.
- We need to translate the outcomes into an expectation value or a _cost function_. A _cost function_ is a problem-specific function that defines the problem's outcome as something to be minimized or maximized, and it's typically used for variational quantum algorithms with various applications in chemistry, machine learning, finance, optimization, etc.
- We should keep in mind the limitations and advantages of the hardware that we are going to use. In particular, the backend's connectivity is important: we want to map the problem to a circuit that is as close as possible to the IBM quantum processors heavy-hex topology to avoid having to implement a lot of SWAP gates.

Once the initial state is prepared, we need to build a variational ansatz for the VQC. The ansatz typically consists of a collection of parametrized rotation gates and entangling layers which are repeated a given number of times. By tuning the parameters of the rotations, we can explore a specific area of the Hilbert space spanned by our five qubits. 

Choosing the right ansatz that allows us to converge to the desired solution is not an easy problem. In general, there are a few trade-offs to consider:

- Speed: By reducing the search space, and thus the number of gates and the depth of the ansatz, the algorithm can run faster.
- Accuracy: Reducing the search space could risk excluding the actual solution to the problem, leading to suboptimal solutions.
- Noise: Deeper circuits are affected by noise, so we need to experiment with our ansatz's connectivity, gates, and gate fidelity.

### QP steps 2&3: Optimize problem for quantum execution and execute using Qiskit primitives

In this step, we will use the Qiskit's transpiler to map our circuit onto an actual, physical qubit hardware, to then execute it using the Estimator primitive.

- [Transpilation](https://docs.quantum.ibm.com/transpile) includes mapping each qubit of our circuit to an actual qubit in the device and then translating the individual quantum gates into sequences of operations that are performed on the hardware. In certain cases, SWAP gates need to be added in order to adapt the connectivity to the topology of the devices, which follow a [heavy-hex pattern](https://www.ibm.com/quantum/blog/heavy-hex-lattice#:~:text=The%20heavy%2Dhex%20lattice%20represents,additional%20qubit%20on%20each%20edge.). You will learn more about this in the other Labs!

- The [Estimator primitive](https://docs.quantum.ibm.com/api/qiskit-ibm-runtime/qiskit_ibm_runtime.EstimatorV2) uses a complex algebraic process to estimate the expectation value on a real quantum device, by breaking down the observable into a combination of other observables with known eigenbases.

Let's start by defining the cost function for this problem, which will be minimized during the training process in order to learn how to classify the birds
in the right category. In this case, the cost function is defined in terms of the expectation values of an observable $\hat{O}$ on the outputs of the circuit for each of the birds in the dataset.

$C(\theta) = \sum_{i\in \text{birds}} (\langle \psi_i(\theta) \vert \hat{O} \vert \psi_i(\theta) \rangle - L_i),$

where $\psi_i(\theta)$ is the output state of the circuit for the bird $i$ and $L_i$ is the label for the same bird. Here, $\theta$ is the vector of parameters for the ansatz.

If we successfully minimize our cost function, the VQC will output an expectation value of $\langle ZZZZZ \rangle = 1$ for IBM-quantum birds and $\langle ZZZZZ \rangle = 0$ for non-IBM quantum birds.

````python
# Define observable 
obs = SparsePauliOp("ZZZZZ")
````

````python
# Using StatevectorEstimator from qiskit primitives (v2) to get exact results for now
estimator = StatevectorEstimator()
aer_sim = AerSimulator(method='statevector')
pm = generate_preset_pass_manager(backend=aer_sim, optimization_level=1, seed_transpiler=0)

# Define cost function
def cost_func(params):

    cost = 0
    for amplitudes,label in zip(list_coefficients, list_labels):
        qc = QuantumCircuit(num_qubits)
        # Amplitude embedding
        qc.initialize(amplitudes)
        # Initial state + ansats
        classifier = qc.compose(ansatz.decompose(reps=6)) 
        # Transpile classifier
        transpiled_classifier = pm.run(classifier) 
        # Run estimator
        pub = (transpiled_classifier, obs, params)
        job = estimator.run([pub])
        # Get result
        result = job.result()[0].data.evs
        # Compute cost function
        cost += np.abs(result - label)
        
    return cost
````

### QP step 4: Classical post-processing stage

The last step of the workflow includes any classical post-processing that we might implement on the output. For the case of variational quantum algorithms, this stage includes the classical optimization over the ansatz parameters in order to minimize the cost function we have defined previously. This is the part in which we **TRAIN** the VQC.

Note that we need to define an initial set of parameters for the ansatz. The initial point of the minimization is very important, since being far away from the minimum can really complicate the convergence process. Because of this, we will use five pre-saved configurations of initial parameters for the VQC, which can be found in _params_0.npy_. 


````python
# Intialize lists to store results from different runs
cost_history_list = []
res_list = []

# Retrieve initial parameters
params_0_list = np.load('params_0.npy')

for it in range(5):

    params_0 = params_0_list[it]
    cost_history = []
    # Minimize the cost function
    res = minimize(cost_func, params_0, method="COBYLA", 
                   callback=lambda params: cost_history.append(cost_func(params)), 
                   options={'maxiter':200}) ## Add 
    
    # Print results after convergence
    print(res)

    # Save results from different runs
    res_list.append(res)
    cost_history_list.append(cost_history)
````

## Part B: Build your own noisy backend to test the VQC 

We have seen how we can perform the 4 steps in general, now we want to test it if that still works if we add noise. For this lab we focus on local testing, thus we want to test on our local machine with noise. For this reason we now want to create a fake backend with noise but first lets talk a bit about different types of noise. There are several types:

1. Decoherence: Qubits will lose their information of the quantum state over time, especially if they stay idle after they are initialized. For decoherence we normally don't use an error rate, but instead use T1 and T2 time, the amount of time it takes until a qubit loses it's information.
2. Measurement Errors: When qubits are measured, it can cause errors, meaning that instead of a 0 a 1 is measured and vice versa. This works similar to a classical channel.
3. Gate Errors: Gates are not perfect and have a small chance to introduce an error when applied. This is especially true for 2 qubit gates, like the CX, the CZ or the ECR gate, which normally have roughly a 10 times higher error rate than single qubit gates.
4. Cross Talk Errors: When applying a gate to a qubit, other qubits, especially neighbouring ones, can also be influenced. This is even ore the case if these qubits lie idle. Fortunately on the newest Heron devices, this is less of a problem, but it is still something one should be aware of.

We can see the coherence times as well as the error rates of quantum devices directly on the IBM Quantum page when inspecting a device. Here a short overview:

After we have now learned about noise, we want to make some noise, by using Qiskit to create a fake backend. 

For this, we will make use of the [Generic Backendv2 class](https://docs.quantum.ibm.com/api/qiskit/qiskit.providers.fake_provider.GenericBackendV2), which allows us to specify the noise levels and the connectivity of the backend we want to simulate. 

````python
from qiskit.transpiler import InstructionProperties
from qiskit.providers.fake_provider import GenericBackendV2
 
fake_backend = GenericBackendV2(
    num_qubits=5,
    basis_gates=["id", "rz", "sx", "x", "cx"],
    control_flow=True,
  )
````

If we could only use the fake backend as given that would be not that useful, since in the end we want to create a fake backend which behaves as similarily as possible to a specific backend on which we want to run our Quantum Circuit in the end.

Fortunately we can change the properties of the fake backend we just created, and thus can adapt it to our needs. We will now decrease our error rate to 1e-8 and further we also change the duration of the gates to the same number:

````python
# Updating properties of fake backend
for i in range(fake_backend.num_qubits):
    qarg = (i,)
    fake_backend.target.update_instruction_properties('rz', qarg, InstructionProperties(error=1e-8, duration=1e-8))
    fake_backend.target.update_instruction_properties('x', qarg, InstructionProperties(error=1e-8, duration=1e-8))
    fake_backend.target.update_instruction_properties('sx', qarg, InstructionProperties(error=1e-8, duration=1e-8))
    fake_backend.target.update_instruction_properties('measure', qarg, InstructionProperties(error=1e-8, duration=1e-8))

# Update 2-qubit gate properties (CX gate) for all edges in the chosen coupling map
cx_props = {}
coupling_map = fake_backend.coupling_map
for edge in coupling_map:
    fake_backend.target.update_instruction_properties('cx', tuple(edge), InstructionProperties(error=1e-8, duration=1e-8))
````

# Part C: Apply ZNE to mitigate gate errors

In this part of the Lab, we will use the zero noise extrapolation error (ZNE) technique to mitigate the effect of errors in the calculation of expectation values.

For this, we are interested in using the [_prototype-zne_](https://github.com/qiskit-community/prototype-zne/) module, which will allow us to implement a highly customizable workflow to target our specific problem. Feel free to explore the details of the module's documentation, which also includes short tutorials.



Implementing ZNE includes two phases:

1. Amplification of the noise introduced by the gates in the circuit. This can be done using different techniques, such as ```GlobalFoldingAmplifier```, ```LocalFoldingAmplifier```, ```CxAmplifier```or by creating a customized amplifier targeting our transpiled circuit.
2. Extrapolation of the returned expectation values to the zero noise limit. For this, we can choose a number of noise factors and the fit that we want to use for the extrapolation. For example, the fit could be linear, polynomial (with different degrees), exponential or further customized.
